## Concatenando em df_geral

```python
dfs = [df_2004_1, df_2004_2, df_2005_1, df_2005_2, df_2006_1, df_2006_2, df_2007_1, df_2007_2,
       df_2008_1, df_2008_2, df_2009_1, df_2009_2, df_2010_1, df_2010_2, df_2011_1, df_2011_2,
       df_2012_1, df_2012_2, df_2013_1, df_2013_2, df_2014_1, df_2014_2, df_2015_1, df_2015_2,
       df_2016_1, df_2016_2, df_2017_1, df_2017_2, df_2018_1, df_2018_2, df_2019_1, df_2019_2,
       df_2020_1, df_2020_2, df_2021_1, df_2021_2, df_2022_1, df_2022_2, df_2023_1, df_2023_2]

nome_novo_da_primeira_coluna = 'Região'

for df in dfs:
    nome_atual_da_primeira_coluna = df.columns[0]
    df.rename(columns={nome_atual_da_primeira_coluna: nome_novo_da_primeira_coluna}, inplace=True)

df_geral = pd.concat(dfs, ignore_index=True)

df_geral.head()
```

```python
df_geral.info()
```

### Tratando as colunas de valor (vírgulas para float)

```python
df_geral['Valor de Compra'] = df_geral['Valor de Compra'].astype(str)
df_geral['Valor de Compra'] = df_geral['Valor de Compra'].str.replace(',', '.')
df_geral['Valor de Compra'] = df_geral['Valor de Compra'].astype(float)

df_geral['Valor de Venda'] = df_geral['Valor de Venda'].astype(str)
df_geral['Valor de Venda'] = df_geral['Valor de Venda'].str.replace(',', '.')
df_geral['Valor de Venda'] = df_geral['Valor de Venda'].astype(float)
```

```python
### Retirando colunas que não serão utilizadas

df_geral = df_geral.drop(['Revenda', 'CNPJ da Revenda', 'Nome da Rua', 'Numero Rua', 'Complemento', 'Bairro', 'Cep', 'Valor de Compra', 'Unidade de Medida', 'Bandeira'], axis=1)
```

```python
# Passo 1: Agrupar por 'Produto' e por mês em 'Data da Coleta', e calcular o preço médio
df_geral['Mes'] = df_geral['Data da Coleta'].dt.to_period('M')  # Adiciona uma coluna 'Mes' com o mês e ano da coleta
preco_medio_mensal = df_geral.groupby(['Produto', 'Mes'])['Valor de Venda'].mean().reset_index()

# Passo 2: Pivotar para ter produtos como colunas e meses como linhas
df_pivot = preco_medio_mensal.pivot(index='Mes', columns='Produto', values='Valor de Venda')

# Converte o índice de Period para datetime (primeiro dia de cada mês)
df_pivot.index = df_pivot.index.to_timestamp()

# Encontrando os anos únicos no seu DataFrame
anos_unicos = pd.to_datetime(df_pivot.index).year.unique()
# Criando ticks para o eixo x para o primeiro dia de cada ano único
ticks_x = pd.to_datetime(['{}-01-01'.format(ano) for ano in anos_unicos])

# Plotagem
plt.figure(figsize=(12, 8))
for produto in df_pivot.columns:
    plt.plot(df_pivot.index, df_pivot[produto], linestyle='-', label=produto)

plt.title('Evolução do Preço Médio Mensal por Produto')
plt.xlabel('Data')
plt.ylabel('Preço Médio')
plt.xticks(ticks_x, labels=anos_unicos, rotation=45)  # Configurando os ticks do eixo x para os anos
plt.legend()
plt.tight_layout()
plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='grey', axis='both')
max_value = df_pivot.max().max() + 0.5
ticks_y = np.arange(0, max_value, 0.5)
plt.yticks(ticks_y)

plt.show()
```

### Formatando os dados 

```python
df_dolar['Valor Dólar'] = df_dolar['Valor Dólar'].astype(str)
df_dolar['Valor Dólar'] = df_dolar['Valor Dólar'].str.replace(',', '.')
df_dolar['Valor Dólar'] = df_dolar['Valor Dólar'].astype(float)

df_inflacao['IPCA acumulado % a.a'] = df_inflacao['IPCA acumulado % a.a'].astype(str)
df_inflacao['IPCA acumulado % a.a'] = df_inflacao['IPCA acumulado % a.a'].str.replace(',', '.')
df_inflacao['IPCA acumulado % a.a'] = df_inflacao['IPCA acumulado % a.a'].astype(float)

df_sal_minimo['Valor Sal Minimo'] = df_sal_minimo['Valor Sal Minimo'].astype(str)
df_sal_minimo['Valor Sal Minimo'] = df_sal_minimo['Valor Sal Minimo'].str.replace(',', '.')
df_sal_minimo['Valor Sal Minimo'] = df_sal_minimo['Valor Sal Minimo'].astype(float)

df_inflacao['Data da Coleta'] = df_inflacao['Data da Coleta'].astype(str)
df_inflacao['Data da Coleta'] = df_inflacao['Data da Coleta'].str.replace('.', '/')
```

```python
# Resetando o índice de df_fatores e convertendo o índice anterior em uma coluna
df_fatores_reset = df_fatores.reset_index()

# Resetando o índice de df_combustivel_brasil e convertendo o índice anterior em uma coluna
df_combustivel_brasil_reset = df_combustivel_brasil.reset_index()
```

```python
df_final = pd.concat([df_combustivel_brasil_reset, df_fatores_reset], axis=1)
df_final
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.graph_objs as go
import plotly.express as px
import pandas as pd
import geopandas as gpd
import geobr
import json
from dash import Input, Output, dcc, html
from plotly.subplots import make_subplots


estados = geobr.read_state()
estados['id'] = estados.index.astype(str)  # Adiciona uma coluna 'id' para uso no Plotly

app = dash.Dash(__name__)

# Layout do aplicativo Dash
app.layout = html.Div([
    #html.H1('Análise de Dados Integrada: Impacto dos Combustíveis, Variação do Dólar e Indicadores Econômicos no Brasil',
            #style={'textAlign': 'center', 'color': 'white', 'padding': '30px', 'margin-left': '40px', 'margin-right': '40px'}),

    # Contêiner cinza arredondado
    html.Div([
        #html.H1('Análise de Dados Integrada: Impacto dos Combustíveis, Variação do Dólar e Indicadores Econômicos no Brasil',
            #style={'textAlign': 'center', 'color': 'white', 'padding': '30px', 'margin-left': '40px', 'margin-right': '40px'}),
        
        html.H1('Impacto dos Fatores Socioeconômicos no Preço dos Combustíveis no Brasil: Uma Análise Integrada',
            style={
                'textAlign': 'center',
                'color': 'white',
                'padding': '30px',
                'margin-left': '40px',
                'margin-right': '40px',
                'font-size': '34px', 
                'text-shadow': '2px 2px 4px #000000',  # Adiciona sombra ao texto
                #'font-family': 'Arial, sans-serif',  
                #'background-color': '#3B5998',  
                'border-radius': '10px' 
            }
        ),
        
        
        html.Div([

            # Título e Filtro de Data
            html.Div([
                html.Label('Selecione o intervalo de datas:', style={'color': 'black', 'margin-bottom': '5px'}),
                dcc.DatePickerRange(
                    id='data-picker-range',
                    start_date=df_geral['Data da Coleta'].min(),
                    end_date=df_geral['Data da Coleta'].max(),
                    display_format='YYYY-MM-DD',
                    style={'width': '100%'}
                )
            ], style={'flex': '1', 'display': 'flex', 'flex-direction': 'column', 'padding': '10px'}),

            # Título e Filtro de Produto
            html.Div([
                html.Label('Selecione o(s) produto(s):', style={'color': 'black', 'margin-bottom': '5px'}),
                dcc.Dropdown(
                    id='produto-dropdown',
                    options=[{'label': produto, 'value': produto} for produto in df_geral['Produto'].unique()],
                    value=df_geral['Produto'].unique().tolist(),  # Todos os produtos são selecionados por padrão
                    clearable=False,
                    multi=True,
                    style={'width': '100%'}
                )
            ], style={'flex': '1', 'display': 'flex', 'flex-direction': 'column', 'padding': '10px'})
        ], style={'display': 'flex', 'background-color': 'white', 'border-radius': '10px', 'padding': '20px', 'margin': '20px', 'margin-top': '50px'}),

    # Gráficos e outros elementos dentro do contêiner cinza
 
        html.Div([
            dcc.Graph(id='preco-medio-mensal-graph',style={'background-color': 'black', 'border-radius': '10px'}),
        ], style={'width': '48%', 'display': 'inline-block','margin-top': '20px', 'vertical-align': 'top', 'margin-right': '10px', 'margin-left': '20px'}),

        html.Div([
            dcc.Graph(id='dolar-combustiveis-graph',style={'background-color': 'white', 'border-radius': '10px', 'padding': '0px'}),
        ], style={'width': '48%', 'display': 'inline-block', 'margin-top': '20px', 'vertical-align': 'top', 'margin-right': '10px', 'margin-left': '10px'}),

        html.Div([
            dcc.Graph(
            id='sal-minimo-combustivel-graph',
            style={'height': '350px', 'background-color': 'white', 'border-radius': '10px', 'padding': '0px'}  # Define a altura desejada
        ),
        ], style={'width': '97%', 'display': 'inline-block', 'margin-right': '10px', 'margin-top': '20px', 'vertical-align': 'top','margin-left': '20px'}),

        html.Div([
            dcc.Graph(
            id='valor-dolar-graph',
            style={'height': '260px', 'background-color': 'white', 'border-radius': '10px', 'padding': '0px'}  # Define a altura desejada
        ),
            dcc.Graph(
            id='ipca-acumulado-graph',
            style={'height': '260px', 'background-color': 'white', 'border-radius': '10px', 'padding': '0px'}  # Define a altura desejada
        ),
        ], style={'width': '48%', 'display': 'inline-block', 'margin-right': '10px', 'margin-top': '20px', 'vertical-align': 'top', 'margin-left': '20px'}),

        html.Div([
            dcc.Graph(
            id='mapa-calor-estados',
            style={'height': '520px', 'background-color': 'white', 'padding': '0px'}  # Define a altura desejada
        ),
        ], style={'width': '48%', 'display': 'inline-block', 'margin-right': '10px', 'margin-top': '20px', 'vertical-align': 'top', 'margin-left': '10px'}),

    ], style={'padding': '20px', 'background': 'linear-gradient(135deg, #5e789c, #3c1053)', 'border-radius': '10px', 'margin': '10px'}),

], style={'backgroundColor': '#272727', 'padding': '20px'})



# Callbacks para os gráficos de preço, valor do dólar, etc.
@app.callback(
    Output('preco-medio-mensal-graph', 'figure'),
    [Input('produto-dropdown', 'value'),
     Input('data-picker-range', 'start_date'),
     Input('data-picker-range', 'end_date')]
)
def update_line_chart(produtos_selecionados, start_date, end_date):
    df_filtrado = df_final.loc[start_date:end_date, produtos_selecionados].copy()
    fig = go.Figure()
    for produto in produtos_selecionados:
        fig.add_trace(go.Scatter(x=df_filtrado.index, y=df_filtrado[produto], mode='lines', name=produto))
    fig.update_layout(title='Evolução do Preço Médio dos Combustíveis',
                      xaxis_title='Data',
                      yaxis_title='Preço Médio (R$)',
                      legend_title='Produto',
                      colorway=px.colors.qualitative.Dark2,
                      margin=dict(l=80, r=80, t=90, b=80),
                      #plot_bgcolor='rgba(238, 238, 238, 0.8)',
                      #paper_bgcolor='#eeeeee'  # Fundo transparente
    )
    return fig

@app.callback(
    Output('dolar-combustiveis-graph', 'figure'),
    [Input('produto-dropdown', 'value'), Input('data-picker-range', 'start_date'), Input('data-picker-range', 'end_date')]
)
def update_scatter_chart(produtos_selecionados, start_date, end_date):
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)

    # Filtrar os dados com base no intervalo de datas
    df_filtrado = df_final.loc[(df_final.index >= start_date) & (df_final.index <= end_date)]
    df_filtrado.reset_index(inplace=True)  # Reseta o índice para transformar 'Data da Coleta' em uma coluna
    df_filtrado['Ano'] = df_filtrado['Data da Coleta'].dt.year

    # Preparando dados para o Plotly Express, convertendo o dataframe em formato longo
    df_long = pd.melt(df_filtrado, id_vars=['Data da Coleta', 'Ano', 'Valor Dólar'], value_vars=produtos_selecionados,
                      var_name='Produto', value_name='Preço')

    # Limpar dados e converter para numérico onde necessário
    df_long.dropna(subset=['Preço', 'Valor Dólar'], inplace=True)
    df_long['Preço'] = pd.to_numeric(df_long['Preço'], errors='coerce')
    df_long['Valor Dólar'] = pd.to_numeric(df_long['Valor Dólar'], errors='coerce')
    df_long['Tamanho'] = 15

    # Criar o gráfico de dispersão animado por ano
    fig = px.scatter(df_long,
                     x="Preço",  
                     y="Valor Dólar", 
                     color="Produto",
                     size="Tamanho",  
                     hover_name="Produto",
                     animation_frame="Ano",
                     #size_max=40,
                     range_x=[df_long['Preço'].min()*0.9, df_long['Preço'].max()*1.1],
                     range_y=[df_long['Valor Dólar'].min()*0.9, df_long['Valor Dólar'].max()*1.1],
                     title="Evolução do Preço dos Combustíveis vs. Valor do Dólar",
                     color_discrete_sequence=px.colors.qualitative.Dark2) 

    fig.update_layout(transition={'duration': 500},
                    xaxis_title="Preço Combustível", 
                    yaxis_title="Valor Dólar",  
                    margin=dict(l=80, r=80, t=90, b=100) 
                    )

    return fig
    

@app.callback(
    Output('sal-minimo-combustivel-graph', 'figure'),
    [Input('produto-dropdown', 'value'),
     Input('data-picker-range', 'start_date'),
     Input('data-picker-range', 'end_date')]
)
def update_sal_minimo_combustivel_graph(produtos_selecionados, start_date, end_date):
    # Converter as datas de início e término para datetime
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)

    # Filtrar os dados com base no intervalo de datas
    df_filtrado = df_final.loc[(df_final.index >= start_date) & (df_final.index <= end_date)]

    fig = go.Figure()
    for produto in produtos_selecionados:
        # Garantir que a coluna existe para evitar erros
        if f'Sal Minimo / {produto}' in df_filtrado.columns:
            fig.add_trace(go.Scatter(
                x=df_filtrado.index, 
                y=df_filtrado[f'Sal Minimo / {produto}'],
                mode='lines', 
                name=f'{produto}'
            ))

    fig.update_layout(
    title='Poder de Compra de Combustível<br><sub>(Salário mínimo/Preço combustível)</sub>',
    xaxis_title='Data',
    yaxis_title='Litros de combustível',
    legend_title='Produto',
    colorway=px.colors.qualitative.Dark2
    )

    return fig





@app.callback(
    Output('valor-dolar-graph', 'figure'),
    [Input('produto-dropdown', 'value'),
     Input('data-picker-range', 'start_date'),
     Input('data-picker-range', 'end_date')]
)
def update_valor_dolar_graph(produtos_selecionados, start_date, end_date):
    # Converter datas de início e término para datetime
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    
    # Verificar se as datas são válidas
    if start_date is None or end_date is None:
        return go.Figure()  # Retorna um gráfico vazio se as datas não forem válidas
    
    # Filtrar os dados com base no intervalo de datas
    df_filtrado = df_final.loc[(df_final.index >= start_date) & (df_final.index <= end_date)]
    
    # Verificar se o DataFrame filtrado não está vazio
    if df_filtrado.empty:
        return go.Figure()  # Retorna um gráfico vazio se não houver dados no intervalo selecionado
    
    # Criando o gráfico de linha
    fig = px.line(df_filtrado, x=df_filtrado.index, y='Valor Dólar', title='Valor Dólar',
                  color_discrete_sequence=['#66a61f'])
    
    # Adicionando título e rótulos dos eixos
    fig.update_layout(
        title='Valor do Dólar ao Longo do Tempo',
        title_font_size=16,
        xaxis_title='Data',
        yaxis_title='Valor do Dólar ($)'
    )
    
    # Adicionando retângulos verticais para marcar os eventos específicos dentro do intervalo filtrado
    eventos = {
        "Crise Financeira Internacional": {'start_date': '2007-01-01', 'end_date': '2008-12-31', 'color': 'orange'},
        "Pandemia COVID-19": {'start_date': '2020-03-01', 'end_date': '2021-12-31', 'color': 'blue'},
        "Crise Econômica Brasileira": {'start_date': '2014-01-01', 'end_date': '2016-12-31', 'color': 'red'}
    }
    
    for evento, data in eventos.items():
        evento_start_date = pd.to_datetime(data['start_date'])
        evento_end_date = pd.to_datetime(data['end_date'])
        
        # Verificar se o evento está dentro do intervalo filtrado
        if evento_start_date <= end_date and evento_end_date >= start_date:
            # Ajustar as datas do evento para o intervalo filtrado
            x0 = max(evento_start_date, start_date)
            x1 = min(evento_end_date, end_date)
            
            fig.add_vrect(
                x0=x0, x1=x1,
                fillcolor=data['color'], opacity=0.3,
                layer="below", line_width=0,
                annotation_text=evento,
                annotation_position="top left",
                annotation_font=dict(
                    size=11,
                    color="black"
                ),
                annotation_bgcolor="rgba(255, 255, 255, 0.7)"  # Fundo branco com opacidade
            )
    
    # Ajusta os limites do eixo y para melhorar a visualização
    fig.update_yaxes(range=[0, df_filtrado['Valor Dólar'].max() * 1.4])  # Aumenta o eixo y
    
    # Ajusta o layout para mostrar todas as anotações e legendas
    fig.update_layout(showlegend=True, font=dict(size=10))
    return fig




@app.callback(
    Output('ipca-acumulado-graph', 'figure'),
    [Input('produto-dropdown', 'value'),
     Input('data-picker-range', 'start_date'),
     Input('data-picker-range', 'end_date')]
)
def update_ipca_acumulado_graph(produtos_selecionados, start_date, end_date):
    # Converter datas de início e término para datetime
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)

    # Filtrar os dados com base no intervalo de datas
    df_filtrado = df_final.loc[(df_final.index >= start_date) & (df_final.index <= end_date), 'IPCA acumulado % a.a']

    # Criando o gráfico de linha
    fig = px.line(df_filtrado, x=df_filtrado.index, y=df_filtrado, title='IPCA acumulado % a.a.', 
                  color_discrete_sequence=['#66a61f'])

    # Datas e eventos para anotar
    events = {
        '2002-01-01': 'Lula',
        '2011-01-01': 'Dilma Rousseff',
        '2016-09-01': 'Michel Temer',
        '2019-01-01': 'Jair Bolsonaro',
        '2023-01-01': 'Luiz Inácio Lula',
    }

    # Adicionando anotações com verificação de existência da data
    for date, event in events.items():
        if pd.Timestamp(date) in df_filtrado.index:
            y_value = df_filtrado.loc[pd.Timestamp(date)]
            fig.add_annotation(
                x=pd.Timestamp(date), y=y_value, 
                text=event, showarrow=True, arrowhead=1, 
                ax=0, ay=-40, bgcolor="rgba(255, 255, 255, 0.7)",
                font=dict(size=11)
            )

    # Adicionando título e rótulos dos eixos
    fig.update_layout(
        title='Variação do IPCA Acumulado ao Longo do Tempo',
        title_font_size=16,
        xaxis_title='Data',
        yaxis_title='IPCA Acumulado (% a.a.)',
        xaxis=dict(
            title_font=dict(size=12),
            tickfont=dict(size=10)
        ),
        yaxis=dict(
            title_font=dict(size=12),
            tickfont=dict(size=10)
        )
    )
    # Ajusta o layout para mostrar todas as anotações e evitar sobreposição
    fig.update_yaxes(range=[df_filtrado.min() - 1, 20])  # Aumentando o limite superior do eixo y
    fig.update_layout(showlegend=True)

    return fig




@app.callback(
    Output('mapa-calor-estados', 'figure'),
    [Input('produto-dropdown', 'value'),
     Input('data-picker-range', 'start_date'),
     Input('data-picker-range', 'end_date')]
)
def update_map(produtos_selecionados, start_date, end_date):
    df_filtrado = df_geral[df_geral['Produto'].isin(produtos_selecionados)]
    df_filtrado['Data da Coleta'] = pd.to_datetime(df_filtrado['Data da Coleta'])
    
    df_filtrado = df_filtrado[(df_filtrado['Data da Coleta'] >= pd.to_datetime(start_date)) &
                              (df_filtrado['Data da Coleta'] <= pd.to_datetime(end_date))]
    
    media_venda_por_estado = df_filtrado.groupby(['Estado', 'Produto'])['Valor de Venda'].mean().reset_index()
    
    if media_venda_por_estado.empty:
        return go.Figure()

    estados = geobr.read_state()
    estados['id'] = estados.index.astype(str)

    geojson = json.loads(estados.geometry.to_json()) if isinstance(estados.geometry, gpd.GeoSeries) else estados.geometry

    fig = make_subplots(
        rows=2, cols=2,
        specs=[[{"type": "choropleth"}, {"type": "choropleth"}],
               [{"type": "choropleth"}, {"type": "choropleth"}]],
        subplot_titles=produtos_selecionados,
        horizontal_spacing=0.02,  
        vertical_spacing=0.05    
    )

    zmin = media_venda_por_estado['Valor de Venda'].min()
    zmax = media_venda_por_estado['Valor de Venda'].max()

    for i, produto in enumerate(produtos_selecionados, start=1):
        pos = ((i-1)//2 + 1, (i-1)%2 + 1)
        df_produto = media_venda_por_estado[media_venda_por_estado['Produto'] == produto]
        if df_produto.empty:
            continue

        estados_merged = estados.merge(df_produto, how='left', left_on='abbrev_state', right_on='Estado')
        estados_merged['text'] = estados_merged['name_state'] + '<br>' + \
                                 'Preço Médio: ' + estados_merged['Valor de Venda'].round(2).astype(str) + '<br>' + \
                                 'Produto: ' + estados_merged['Produto']

        fig.add_trace(
            go.Choropleth(
                geojson=geojson,
                locations=estados_merged['id'],
                z=estados_merged['Valor de Venda'],
                text=estados_merged['text'], 
                hoverinfo='text+z+location',  
                colorscale="Magma",
                zmin=zmin,
                zmax=zmax,
                marker_line_color='white',
                marker_line_width=0.5,
                colorbar_title="Preço Médio"
            ), 
            row=pos[0], col=pos[1]
        )

    fig.update_geos(fitbounds="locations", visible=True)
    fig.update_layout(
        margin={"r":10, "t":90, "l":15, "b":0},
        title_text="Análise de Preços por Estado e Produto",
        height=520,
        width=850
    )
    # Ajustando a posição dos títulos dos subplots
    for i, annotation in enumerate(fig['layout']['annotations']):
        if i < 2:  # Ajusta apenas os dois primeiros títulos (superiores)
            annotation['y'] -= 0.03  # Aumenta o valor de 'y' para subir os títulos
        else:  # Ajusta os dois últimos títulos (inferiores)
            annotation['y'] -= 0.03  # Diminui o valor de 'y' para descer os títulos

    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```
